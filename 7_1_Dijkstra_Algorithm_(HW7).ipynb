* BFS를 사용하여 최소거리 경로를 찾을 수 있다.  
* 최소 cost (시간, 신호등 수, 주변 경관, $\cdots$)를 가지는 경로를 찾을 수 있는 알고리즘  
* 탐욕 알고리즘 중 하나
* 첫 정점을 기준으로 연결되어 있는 정점들을 추가해가며, 최단 거리를 갱신하는 것 (정점 연결 전에는 시작점을 제외한 정점들은 모두 무한대 값을 가진다.)

    1. nodes에서 최소거리 노드를 찾는다. 시작점 거리는 0, 나머지는 infinity로 설정한다. (시작점이 현재 노드)
    2. 현재 노드를 방문 처리
    3. 현재 노드의 이웃 노드를 구하고 시작점에서 현재 노드를 통해 이웃 노드로 가는 비용과 이 노드를 거치지 않고 가는 비용을 비교하여 현재 노드를 통해 가는 것이 작으면 비욕과 부모 노드 업데이트
    4. 현재 노드는 노드 집합에서 삭제, 방문 리스트에 삽입하고 방문하지 않은 노드들 중에 최소 비용 노드를 찾아 현재 노드로 지정하고 2로 이동한다.  

* ADT
    1. 그래프 경로 설정
    2. 노드 집합 만들기 $\Rightarrow$ visits로 옮기면 알고리즘 끝 (nodes: set, visits: set)
    3. 출발점을 기준으로 cost 설정 (출발은 0, 나머지는 inf, 부모는 None인상태) (cost: dictionary, curNode=start)
    4. nodes에서 최소 비용을 가지는 노드를 curNode로 설정 (start)
    5. 해당 노드에서 연결된 노드들을 찾고 cost를 갱신하여 반환한다. (curNode에서 갈 수 있는 노드를 return 한다.) (def _neighbor) (neighbor: dictionary)
    6. 인접한 두 노드의 가중치를 리턴하는 함수 (_getWeight(n1, n2))  
    neighbor에서 노드를 꺼내서 해당 노드와 curNode의 가중치와 cost(출발점으로부터의 거리)를 비교하여 curNode를 통해 가는 것이 저렴하면 cost 값을 갱신하고 부모 노드를 curNode로 갱신한다.  
    7. curNode를 visits로 이동하고 nodes에서 삭제한다.
    8. 남아있는 nodes 중 최소 비용을 가지는 노드를 curNode로 선정한다. (cost에서 nodes에 있는 key만 선택한다.) (dicfilt 코드 참조)
    9. curNode의 이웃에 대해 6~7을 수행
    10. nodes에 아무것도 남지 않을 때까지 알고리즘을 반복한다.

```
import csv

f = open('/gdrive/My Drive/Python/알고리즘/data/subway.csv', 'r', encoding='utf-8-sig')
csvReader = csv.reader(f)

start = []
end = []
time = []
for row in csvReader:
    start.append(row[0])
    end.append(row[1])
    time.append(int(row[2]))
f.close()

# print(start)
# print(end)

sub = tuple([ [start[i], end[i], time[i]] for i in range(len(start)) ])
sub += tuple([ [end[j], start[j], time[j]] for j in range(len(end)) ])
sub
```

```
class Dijkstra:
    def __init__(self, graph):
        self.graph = graph
        
        nodes = set()
        for edge in graph:
            nodes.add(edge[0])
            nodes.add(edge[1])
        self.node = nodes
    
    def _neighbor(self, curNode): # 이웃노드 확인하는 메서드
        neighbor = {}
        for edge in self.graph:
            if edge[0] == curNode:
                neighbor[edge[1]] = edge[2]
            elif edge[1] == curNode:
                neighbor[edge[0]] = edge[2]
        return neighbor
    
    def _getWeight(self, n1, n2):
        for edge in self.graph:
            if edge[0] == n1 and edge[1] == n2:
                return edge[2]
            elif edge[1] == n1 and edge[0] == n2:
                return edge[2]
        return None
    
    def getPath(self, start, end):
        visits = set()
        
        # 초기 cost 설정해주기
        cost = {}
        for node in self.node:
            cost[node] = [float("inf"), "None"]
        cost[start][0] = 0 # 시작점의 cost는 0이다.

        curNode = start
        while True:
            neighbors = self._neighbor(curNode) # curNode에서 갈 수 있는 이웃들 반환
            for node in neighbors: # curNode의 이웃 노드들 중에서 
                if self._getWeight(curNode, node) + cost[curNode][0] < cost[node][0]: # 시작점으로부터 curNode까지의 거리 + 이웃 노드와 curNode의 가중치가 / 시작점으로 부터의 이웃 노드까지의 거리보다 작으면
                    cost[node][0] = self._getWeight(curNode, node) + cost[curNode][0] # 시작점으로부터 이웃노드 까지의 거리를 갱신하고
                    cost[node][1] = curNode # cost에 있는 부모노드도 갱신해라.

            # curNode를 마무리 해준다.
            visits.add(curNode)
            self.node.remove(curNode)
            
            if len(self.node) > 0:
                dictfilt = lambda x, y: dict([ (i,x[i]) for i in x if i in set(y) ])
                curNode = min(dictfilt(cost, self.node), key=dictfilt(cost, self.node).get)
            else:
                break
                
        path = [end]
        time = 0

        while end != start:
            time += self._getWeight(end, cost[end][1])
            path.append(cost[end][1])
            end = cost[end][1]
        return path[::-1], time

subway = Dijkstra(sub)
print(subway.getPath('서울역(1)', '선릉(2)'))
```
