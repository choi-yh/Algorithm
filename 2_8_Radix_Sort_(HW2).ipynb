* 자리수를 기준으로 정렬하는 방법으로 문자열이나 자리수가 같은 자료를 정렬하는데 효과적이다.  

* 가장 낮은 자리수부터 비교하면서 정렬한다.

* 비교 연산을 하지 않으며 정렬 속도가 빠르지만 메모리(기수 테이블 크기의 메모리) 가 많이 필요하다.  

* 시간복잡도는 $O(dn)$ (d: 가장 큰 데이터의 자리수)

* 숫자 예시

    Step 1: 0 ~ 9. 10개의 Queue를 만든다.  

    Step 2: 가장 낮은 자리수부터 자신의 위치에 맞는 Queue에 삽입한다. (Queue니깐 꺼낼 때는 FIFO 방식)

    Step 3: 자리수를 늘려가며 가장 큰 자리수에 도달할 때까지 반복한다.

### 기수정렬을 사용하여 한글 단어 정렬 클래스로 만들기 (오름차순, 내림차순)

```
class Radix:
    def __init__(self, words):
        self.words = words
        
    def _maxlen(self):
        maxlen = 0
        for word in self.words:
            if len(word) > maxlen:
                maxlen = len(word)
        return maxlen
        
    def engSort(self):
        k = 27
        for i in range(1, self._maxlen()+1): # 단어자리수 만큼 루프
            queue = [list() for _ in range(k)] # 기수만큼 루프하면서 빈 리스트 생성 (빈큐 27개 만들기)
            
            for word in self.words: #단어 꺼내면서
                # 자리수 맞추기 위해 스페이스 추가하기
                word += (self._maxlen() - len(word)) * ' '
                # ascii 코드를 이용하여 큐에 푸시할 위치를 찾기. 만약, 스페이스면 0번째에 푸시
                if ord(word[-i]) >= 97: # ord('a') == 97
                    queue[ord(word[-i]) - 96].append(word)
                else:
                    queue[0].append(word)
            
            # 큐에 정렬한 원소들을 원 리스트에 갱신 후 반복
            self.words = sum(queue, [])
            
        # 단어에 추가한 스페이스를 제거
        for j in range(len(self.words)):
            self.words[j] = self.words[j].replace(" ", "")
        return self.words

    def korSort(self):
        k = ord("힣") - ord("가") + 2
        for i in range(1, self._maxlen()+1):
            queue = [list() for _ in range(k)]
            
            for word in self.words:
                word += (self._maxlen() - len(word)) * ' '
                if ord(word[-i]) >= 44032: # ord('가')==44032
                    queue[ord(word[-i]) - 44031].append(word)
                else:
                    queue[0].append(word)
            self.words = sum(queue, [])
        
        for j in range(len(self.words)):
            self.words[j] = self.words[j].replace(" ", "")
        return self.words
    
    def korReverseSort(self):
        k = ord("힣") - ord("가") + 2
        for i in range(1, self._maxlen()+1):
            queue = [list() for _ in range(k)]
            
            for word in self.words:
                word += (self._maxlen() - len(word)) * ' '
                if ord(word[-i]) >= 44032: # ord('가')==44032
                    queue[-(ord(word[-i]) - 44031)].append(word) # 거꾸로 들어가게한다.
                else:
                    queue[0].append(word)
            self.words = sum(queue, [])
        
        for j in range(len(self.words)):
            self.words[j] = self.words[j].replace(" ", "")
        return self.words
```
