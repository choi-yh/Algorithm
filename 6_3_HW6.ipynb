```
import csv

f = open('/gdrive/My Drive/Python/알고리즘/data/subway.csv', 'r', encoding='utf-8-sig')
csvReader = csv.reader(f)

start = []
end = []
time = []
for row in csvReader:
    start.append(row[0])
    end.append(row[1])
    time.append(row[2])
f.close()

print(start)
print(end)
```

```
# 방향성이 있으므로 두번 순환

dic = {}
for i in range(len(start)):
    if start[i] not in dic:
        dic[start[i]] = set([end[i]])
    else:
        tmp = dic[start[i]]
        tmp = tmp.update([end[i]])
        dic[start[i]]
        
for j in range(len(end)):
    if end[j] not in dic:
        dic[end[j]] = set([start[j]])
    else:
        tmp = dic[end[j]]
        tmp = tmp.update([start[j]])
        dic[end[j]]
    
print(dic)
```

```
class BFS:
    def __init__(self, graph, start, end):
        self.graph = graph
        self.start = start
        self.end = end
        
    def minPath(self):
        queue = []
        parentsQueue = []
        queue += self.graph[self.start]
        parentsQueue += [self.start] * len(self.graph[self.start])
        visit = [self.start]
        parents = ['None']
        
        while queue:
            item = queue.pop(0)
            temp = parentsQueue.pop(0)
            if item not in visit:
                queue += self.graph[item]
                parentsQueue += [item] * len(self.graph[item])
                visit.append(item)
                parents.append(temp)
                if item == self.end:
                    visits = dict(zip(visit, parents))
                
        path = [self.end]
        print(visits)
        
        curNode = self.end
        while curNode != 'None':
            curNode = visits[curNode]
            path.append(curNode)
        return path[::-1][1:], len(path)

subway = BFS(dic, '서울역(1)', '선릉(2)')
subway.minPath()
```
